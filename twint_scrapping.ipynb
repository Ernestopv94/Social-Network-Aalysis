```python
import twint
import nest_asyncio
import os
import pandas as pd

# tqdm muestra una barra de progreso con el tiempo estimado que falta
from tqdm import tqdm

# Sin esta línea Twint no funciona
nest_asyncio.apply()
```

```python
# Lee el csv con todos los concejales
df = pd.read_csv('concejales.csv')

#Saca una lista con los concejales
lista = df['username'].values.tolist()
```

```python
filename = 'tweets_concejales.csv'

# Twint no sobreescribe sino que hace append, por lo que hay que borrar el archivo existente
if os.path.exists(filename):
    os.remove(filename)
    
c = twint.Config()

# Obtiene todos los tweets de concejales en los que mencionan a otro concejal
for source in tqdm(lista):
    for target in lista:
        if source != target:
            c.Username = source
            c.Search = '@'+target
            c.since = '2019-06-15' # Debería limitar las fechas pero no funciona bien
            c.Limit = 200 # Debería limitar el número de tweets pero tampoco funciona
            c.Store_csv = True
            c.Output = filename

            twint.run.Search(c)
```
